```python
# Success Metrics

# Intent to try out the advertised streaming service
df_ttest = df.copy()

treatment_condition = ((df_ttest['Treatment_Apple_Music'] == 1) | (df_ttest['Treatment_Spotify'] == 1))
control_condition = ((df_ttest['Control_Apple_Music'] == 1) | (df_ttest['Control_Spotify'] == 1))

# Conditions for observing positive intent for 'try'ing advertised streaming service
try_condition = ((df_ttest['Q18'] == 'Maybe') | (df_ttest['Q18'] == 'Yes') | (df_ttest['Q2039'] == 'Maybe') | (df_ttest['Q2039'] == 'Yes'))

# Conditions for observing positive intent for 'switch'ing to the advertised streaming service
switch_condition = ((df_ttest['Q21'] == 'Maybe') | (df_ttest['Q21'] == 'Yes') | (df_ttest['Q2040'] == 'Maybe') | (df_ttest['Q2040'] == 'Yes'))

# Conditions for change in preference order
# treatment
pref_change_condition_treatment = (((df_ttest['Treatment_Apple_Music'] == 1) & (df_ttest['Q19_2'] > df_ttest['Q19_2.1']) ) | 
                                   ((df_ttest['Treatment_Spotify'] == 1) & (df_ttest['Q19_1'] > df_ttest['Q2041_1'])) )

# control
pref_change_condition_control = (((df_ttest['Control_Apple_Music'] == 1) & (df_ttest['Q19_2'] > df_ttest['Q19_2.1']) ) | 
                                   ((df_ttest['Control_Spotify'] == 1) & (df_ttest['Q19_1'] > df_ttest['Q2041_1'])) )

# Applying conditions
df_ttest['positive_intent_try_treatment'] = (treatment_condition & try_condition).astype(int)
df_ttest['positive_intent_try_control'] = (control_condition & try_condition).astype(int)
df_ttest['positive_intent_switch_treatment'] = (treatment_condition & switch_condition).astype(int)
df_ttest['positive_intent_switch_control'] = (control_condition & switch_condition).astype(int)
df_ttest['positive_intent_pref_treatment'] = (pref_change_condition_treatment).astype(int)
df_ttest['positive_intent_pref_control'] = (pref_change_condition_control).astype(int)

df_ttest['positive_intent_treatment'] = (df_ttest['positive_intent_try_treatment'] | df_ttest['positive_intent_switch_treatment'] | df_ttest['positive_intent_pref_treatment']).astype(int)
df_ttest['positive_intent_control'] = (df_ttest['positive_intent_try_control'] | df_ttest['positive_intent_switch_control'] | df_ttest['positive_intent_pref_control']).astype(int)

df_ttest.iloc[:6,-8:]
```

```python
from scipy.stats import ttest_ind
import numpy as np

treatment_condition = ((df_ttest['Treatment_Apple_Music'] == 1) | (df_ttest['Treatment_Spotify'] == 1))
control_condition = ((df_ttest['Control_Apple_Music'] == 1) | (df_ttest['Control_Spotify'] == 1))

positive_intent_treatment = df_ttest[treatment_condition]['positive_intent_treatment']
positive_intent_control = df_ttest[control_condition]['positive_intent_control']

print('\n****** Statistical Analysis for Positive Effect on Intent of Survey Takers Advertised Streaming Service ******')

print('\n******Summary Stats for Treatment Group******')
print(positive_intent_treatment.describe())
print('\n******Summary Stats for Control Group******')
print(positive_intent_control.describe())
print('\n******Difference in Averages******')
diff = positive_intent_treatment[:-1].mean() - positive_intent_control[:-1].mean()
print(diff)
print('\n******T-Stat and P-Value Respectively******')
t_stat, p_value = ttest_ind(positive_intent_treatment, positive_intent_control, equal_var=False)
print(t_stat, p_value)
```

```python
from scipy.stats import ttest_ind
import numpy as np

treatment_condition = ((df_ttest['Treatment_Apple_Music'] == 1) | (df_ttest['Treatment_Spotify'] == 1))
control_condition = ((df_ttest['Control_Apple_Music'] == 1) | (df_ttest['Control_Spotify'] == 1))

positive_intent_pref_treatment = df_ttest[treatment_condition]['positive_intent_pref_treatment']
positive_intent_pref_control = df_ttest[control_condition]['positive_intent_pref_control']

print('\n****** Statistical Analysis for Intent to CHANGE IN PREFERENCE of the Advertised Streaming Service ******')

print('\n******Summary Stats for Treatment Group******')
print(positive_intent_pref_treatment.describe())
print('\n******Summary Stats for Control Group******')
print(positive_intent_pref_control.describe())
print('\n******Difference in Averages******')
diff = positive_intent_pref_treatment[:-1].mean() - positive_intent_pref_control[:-1].mean()
print(diff)
print('\n******T-Stat and P-Value Respectively******')
t_stat, p_value = ttest_ind(positive_intent_pref_treatment, positive_intent_pref_control, equal_var=False)
print(t_stat, p_value)
```

##### Results
1. Cumulative Positive Intent: 
    Observed Difference between treatment and control: 4.85% (t-stat = 0.4916, insignificant at 90% and 95% Confidence levels)

2. Positive intent to try the advertised streaming service
    Observed Difference between treatment and control: -27% (t-stat = -2.11, significant at 90% and 95% Confidence levels)

3. Positive intent to Switch to the advertised streaming service
    Observed Difference between treatment and control: -23.68% (t-stat = -1.65, insignificant at 90% and 95% Confidence levels)

4. Positive intent measured by change in preference of the advertised streaming service
    Observed Difference between treatment and control: 4.45% (t-stat = 0.5424, insignificant at 90% and 95% Confidence levels)

##### Inference
- The campaign had a statistically significant negative impact on the intent to try the advertised music streaming service, which is a critical concern. It suggests that something about the campaign or the service as presented may have actively discouraged potential users.
- The lack of significant effects on the intent to switch and the change in preference rankings suggests that the campaign did not effectively persuade existing users of other services to consider switching or altering their preferences in favor of the advertised service.
- The overall effectiveness of the advertisement in encouraging positive perceptions or actions towards the music streaming service is questioned, particularly given the significant negative impact on the willingness to try the service. This indicates a need to reassess the campaign's messaging, targeting, or the perceived value of the service offered to better align with potential users' expectations and preferences.
